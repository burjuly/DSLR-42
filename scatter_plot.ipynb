```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/dataset_train.csv', index_col = 0)
```

## What are the two features that are similar ?
## Answer: Astronomy and Defense Against the Dark Arts

```python
fig, ax = plt.subplots()
Gryffindor_x = df['Astronomy'][df['Hogwarts House'] == 'Gryffindor']
Gryffindor_y = df['Defense Against the Dark Arts'][df['Hogwarts House'] == 'Gryffindor']
Slytherin_x = df['Astronomy'][df['Hogwarts House'] == 'Slytherin']
Slytherin_y = df['Defense Against the Dark Arts'][df['Hogwarts House'] == 'Slytherin']
Ravenclaw_x = df['Astronomy'][df['Hogwarts House'] == 'Ravenclaw']
Ravenclaw_y = df['Defense Against the Dark Arts'][df['Hogwarts House'] == 'Ravenclaw']
Hufflepuff_x = df['Astronomy'][df['Hogwarts House'] == 'Hufflepuff']
Hufflepuff_y = df['Defense Against the Dark Arts'][df['Hogwarts House'] == 'Hufflepuff']

ax.scatter(Gryffindor_x, Gryffindor_y, color='yellow', alpha=0.4, marker='*')
ax.scatter(Slytherin_x, Slytherin_y, color='blue', alpha=0.4, marker='*')
ax.scatter(Ravenclaw_x, Ravenclaw_y, color='red', alpha=0.4, marker='*')
ax.scatter(Hufflepuff_x , Hufflepuff_y, color='green', alpha=0.4, marker='*')
plt.xlabel('Astronomy')
plt.ylabel('Defense Against the Dark Arts')
```
