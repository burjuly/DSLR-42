```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/dataset_train.csv', index_col = 0)
```

## Boxplot

```python
df1 = df[['Herbology', 'Defense Against the Dark Arts', 'Divination', 'History of Magic', 'Potions', 'Care of Magical Creatures']]
plt.figure(figsize=(20, 10))
sns.boxplot(df1)
```

```python
df2 = df[['Astronomy', 'Muggle Studies', 'Ancient Runes', 'Transfiguration', 'Charms', 'Flying']]
plt.figure(figsize=(20, 20))
sns.boxplot(df2)
```

## Correlation Matrix

```python
sns.set(font_scale=1.15)
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), cmap='RdBu_r', annot=True, vmin=-1, vmax=1)
```
