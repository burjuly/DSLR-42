```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('datasets/dataset_train.csv', index_col = 0 )
```

```python
df = df[df.columns[5:]]
columns = df.columns
df.describe()
```

```python
df.isnull().sum() 
```

```python
def ft_del_nun(column):
    array = [i for i in column if i == i]
    return array

def ft_count(column):
    return(len(column))

def ft_mean(count, column):
    sum = 0
    for i in column:
        sum += i
    return(round(sum / count, 6))

def ft_std(mean, column, count):
    result = 0
    for i in column:
        result += (i - mean) ** 2
    result =  math.sqrt(result / (count - 1))
    return(round(result, 6))

def ft_min(column):
    result = column[0]
    for i in column:
        if i < result:
            result = i
    return(result)

def ft_max(column):
    result = column[0]
    for i in column:
        if i > result:
            result = i
    return(result)

def ft_percentile(column, count, p):
    column = sorted(column)
    count = (count - 1) * p
    floor = int(np.floor(count))
    ceil = int(np.ceil(count))
    if floor == ceil:
        return(column[int(count)]) 
    else:
        delta = (column[ceil] - column[floor]) * (count - floor) + column[floor]
        return delta
    
def ft_range(max_value, min_value):
    return(max_value - min_value)

def ft_interquartile_range(per_75, per_25):
    return(per_75 - per_25)
    
```

```python
table = pd.DataFrame()
table['describe'] = ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'range', 'interquartile range' ]
table = table.set_index('describe')

for name_column in columns:
    values_column = ft_del_nun(df[name_column])
    count = ft_count(values_column)
    mean = ft_mean(count, values_column)
    std = ft_std(mean, values_column, count) 
    min_value = ft_min(values_column)
    max_value = ft_max(values_column)
    per_50 = ft_percentile(values_column, count, 0.5)
    per_25 = ft_percentile(values_column, count, 0.25)
    per_75 = ft_percentile(values_column, count, 0.75)
    range_ = ft_range(max_value, min_value)
    q_range = ft_interquartile_range(per_75, per_25)
    res = [count, mean, std, min_value, per_25, per_50, per_75, max_value, range_, q_range]
    table[name_column] = res
```
